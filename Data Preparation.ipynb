# Segregating images in closed and open eyes

```python
import os
import shutil
from tqdm import tqdm

Raw_DIR = r'C:\Users\Saptarshi Dey\Downloads\mrlEyes_2018_01\mrlEyes_2018_01'
closed_eyes_dir = r'C:\Users\Saptarshi Dey\Downloads\Prepared_Data\Closed_Eyes'
open_eyes_dir = r'C:\Users\Saptarshi Dey\Downloads\Prepared_Data\Open_Eyes'

for dirpath, dirname, filenames in os.walk(Raw_DIR):
    for i in tqdm([f for f in filenames if f.endswith('.png')]):
        if i.split('_')[4] == '0':
            shutil.copy(os.path.join(dirpath, i), closed_eyes_dir)
        elif i.split('_')[4] == '1':
            shutil.copy(os.path.join(dirpath, i), open_eyes_dir)
```
